```python
import selenium

from selenium import webdriver
PATH = 'C:\Program Files (x86)\chromedriver.exe' #extracting path of the webdriver
driver = webdriver.Chrome(PATH)

from selenium.webdriver.common.keys import Keys
```

```python
import time

time.sleep(5) 
```

```python
driver.get('https://www.myntra.com/shoes')
```

```python
select_price = driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div")
select_price.click()
```

```python
select_color = driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div")
select_color.click()
```

```python
#creating one list which is going to store all the links
shoes_opening_urls = []
url= driver.find_elements_by_xpath('//div[@class= "results-showMoreContainer"]//a')
for i in url[0:3]:
    shoes_opening_urls.append(i.get_attribute('href')) #to retrieve any link we use get_attribute
```

```python
shoes_opening_urls
```

```python
shoes_brand=[]
for i in shoes_opening_urls:
    driver.get(i)
    try:
        brands = driver.find_elements_by_xpath("//h3[@class='product-brand']")
        for i in brands:
            shoes_brand.append(i.text)
    except:
        pass
            
shoes_brand
```

```python
shoes_desc=[]
for i in shoes_opening_urls:
    driver.get(i)
    try:
        description = driver.find_elements_by_xpath("//h4[@class='product-product']")
        for i in description:
            shoes_desc.append(i.text)
    except:
        pass
            
shoes_desc
```

```python
shoes_price=[]
for i in shoes_opening_urls:
    driver.get(i)
    try:
        price = driver.find_elements_by_xpath("//div[@class='product-price']")
        for i in price:
            shoes_price.append(i.text)
    except:
        pass
            
shoes_price
```

```python
print(len(shoes_brand),len(shoes_desc),len(shoes_price))
```

```python
data = list(zip(shoes_brand[0:100],shoes_desc[0:100],shoes_price[0:100]))

import pandas as pd
df = pd.DataFrame(data,columns=['shoes_desc','shoes_desc','shoes_price'])

df
```
