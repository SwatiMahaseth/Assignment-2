```python
#downloading important libraries
import selenium

from selenium import webdriver
PATH = 'C:\Program Files (x86)\chromedriver.exe' #extracting path of the webdriver
driver = webdriver.Chrome(PATH)

from selenium.webdriver.common.keys import Keys
```

Q1: Write a python program to scrape data for “Data Analyst” Job position in
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.

```python
#fetching url of naukri
driver.get('https://www.naukri.com/')

#code to close share your location popup window
popup = driver.find_element_by_id("block")
popup.click()

#giving input for job title as Data Analyst
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')

#giving input for job location as Bangalore
search_loc = driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')

#code to click on search button
search_btn = driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()

#fetching company name for top 10 data
coy_name = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies=[]
for i in coy_name[0:10]:
    companies.append(i.text)

#fetching job title for top 10 data
title = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
job_title=[]
for i in title[0:10]:
    job_title.append(i.text)

#fetching job location for the respective companies   
location = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
company_location=[]
for i in location[0:10]:
    company_location.append(i.text)
    
#fetching ecperience required for the respective companies
experience = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
experience_required=[]
for i in experience[0:10]:
    experience_required.append(i.text)

#creating dataframe
data = list(zip(companies,job_title,company_location,experience_required))

import pandas as pd
df = pd.DataFrame(data,columns=['companies','job_title','company_location','experience_required'])

df.head(10)
```

Q2: Write a python program to scrape data for “Data Scientist” Job position in
“Bangalore” location. You have to scrape the job-title, job-location,
company_name, full job-description. You have to scrape first 10 jobs data.

```python
#fetching url
driver.get('https://www.naukri.com/')
```

```python
#if pop up window occurs
popup = driver.find_element_by_id("block")
popup.click()
```

```python
#entering data into seach field and clicking on search button
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Scientist')

search_loc = driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')

search_btn = driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()

#fetching url to open each job 
job_opening_urls = []
url= driver.find_elements_by_xpath('//a[@class= "title fw500 ellipsis"]')
for i in url[0:10]:
    job_opening_urls.append(i.get_attribute('href')) #to retrieve any link we use get_attribute
```

```python
#handeling this case with exception handeling
import time

job_title=[]
companies=[]
job_description=[]

for i in job_opening_urls:
    driver.get(i)
    time.sleep(5)
    #for job title
    try:
        title = driver.find_element_by_xpath("//h1[@class='jd-header-title']")    
        job_title.append(title.text)
    except:
        job_title.append('-')
     #for company name   
    try:
        coy_name = driver.find_element_by_xpath("//a[@class='pad-rt-8']")    
        companies.append(coy_name.text)
    except:
        companies.append('-')
      #for company description  
    try:
        description = driver.find_element_by_xpath('/html/body/div[1]/main/div[2]/div[2]/section[2]/div[1]')    
        job_description.append(description.text)
    except:
        job_description.append('-')
        
```

```python
#fetching location
location1 = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
company_location1=[]
for i in location1[0:10]:
    company_location1.append(i.text)        
```

```python
#printing length of list
print(len(companies),len(job_title),len(company_location1),len(job_description))
```

```python
#creating dataframe
data = list(zip(companies,job_title,company_location1,job_description))

import pandas as pd
df = pd.DataFrame(data,columns=['companies','job_title','company_location','job_description'])

df.head(10)
```

Q4: Write a python program to scrape data for first 10 job results for Data scientist

ago when job was posted, Rating of the company.

```python
#fetching url of glassdoor
driver.get('https://www.glassdoor.co.in/index.htm')

for i in range(12):

    driver.find_element_by_id('sc.location').send_keys(Keys.BACK_SPACE)

#giving input Data Scientist as job title
search_job = driver.find_element_by_id('sc.keyword')
search_job.send_keys('Data Scientist')

#giving input Noida as job location
search_loc = driver.find_element_by_id('sc.location')
search_loc.send_keys('Noida')

#clicking on search button
search_btn = driver.find_element_by_class_name('css-8zxfjs')
search_btn.click()

#fetching the top 10 data
#top 10 company name
company_name = driver.find_elements_by_xpath("//a[@class=' job-search-key-l2wjgv e1n63ojh0 jobLink']")
companies=[]
for i in company_name[0:10]:
    companies.append(i.text)

#number of days ago job was posted    
no_of_days = driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-17n8uzw']")
no_of_days_ago=[]
for i in no_of_days[0:10] :
    no_of_days_ago.append(i.text)
    
#ratings given to the company
    ratings = driver.find_elements_by_xpath("//span[@class=' job-search-key-srfzj0 e1cjmv6j0']")
company_rating=[]
for i in ratings[0:10] :
    company_rating.append(i.text)
    
#creating dataframe
data = list(zip(companies,no_of_days_ago,company_rating))

import pandas as pd
df = pd.DataFrame(data,columns=['companies','no_of_days_ago','company_rating'])

df.head(10)
```

Q5: Write a python program to scrape the salary data for Data Scientist designation
in Noida location.
You have to scrape Company name, Number of salaries, Average salary, Min
salary, Max Salary.

```python
driver.get('https://www.glassdoor.co.in/Salaries/index.htm')

# clearing job location search_bar

for i in range(12):

    driver.find_element_by_id('LocationSearch').send_keys(Keys.BACK_SPACE)

#entering data into the search field and clicking on search button
    search_job = driver.find_element_by_id('KeywordSearch')
search_job.send_keys('Data Scientist')

search_loc = driver.find_element_by_id('LocationSearch')
search_loc.send_keys('Noida')

search_btn = driver.find_element_by_id('HeroSearchButton')
search_btn.click()

#company name
company_name = driver.find_elements_by_xpath("//a[@class='css-f3vw95 e1aj7ssy3']")
companies=[]
for i in company_name[0:10]:
    companies.append(i.text) 
    
#salary
    salary = driver.find_elements_by_xpath("//span[@class='m-0 css-1b6bxoo']")
salary_record=[]
for i in salary[2:12]:
   salary_record.append(i.text)

#average salary
Average = driver.find_elements_by_xpath("//div[@class='col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")
avg_salary=[]
for i in Average[0:10]:
   avg_salary.append(i.text)

salary_range = driver.find_elements_by_xpath("//span[@class='d-block d-lg-none m-0 css-1b6bxoo']")
min_max_salary_range=[]
for i in salary_range[0:10]:
   min_max_salary_range.append(i.text)

rating = driver.find_elements_by_xpath("//span[@class='m-0 css-kyx745']")
company_rating=[]
for i in rating[0:10]:
   company_rating.append(i.text)

#creating dataframe
data2 = list(zip(companies,salary_record,avg_salary,min_max_salary_range,company_rating))
import pandas as pd
df2 = pd.DataFrame(data2,columns=['companies','salary_record','avg_salary','min_max_salary_range','company_rating'])
df2.head(10)
```

```python
#driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace')

#all_reviews = driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
#all_reviews.click()

review_urls = []
url= driver.find_elements_by_xpath('//nav[@class= "yFHi8N"]//a')
for i in url[0:11]:
    review_urls.append(i.get_attribute('href')) #to retrieve any link we use get_attribute
    
iphone_ratings=[]
for i in review_urls:
    driver.get(i)
    try:
        ratings = driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq' or @class='_3LWZlK _1rdVr6 _1BLPMq']")
        for i in ratings:
            iphone_ratings.append(i.text)
    except:
        pass
    
Review_summary=[]
for i in review_urls:
    driver.get(i)
    try:
        Reviews = driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
        for i in Reviews:
            Review_summary.append(i.text)
    except:
        pass
    
Full_review=[]
for i in review_urls:
    driver.get(i)
    try:
        Review = driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
        for i in Review:
            Full_review.append(i.text)
    except:
        pass
```
