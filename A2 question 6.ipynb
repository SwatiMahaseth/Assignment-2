```python
import selenium

from selenium import webdriver
PATH = 'C:\Program Files (x86)\chromedriver.exe' #extracting path of the webdriver
driver = webdriver.Chrome(PATH)

from selenium.webdriver.common.keys import Keys
```

```python
import time

time.sleep(5) 
```

```python
driver.get('https://www.flipkart.com/')
```

```python
cross_btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
cross_btn.click()
```

```python
search_product = driver.find_element_by_name("q")
search_product.send_keys('Sunglasses')
```

```python
search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
#creating one list which is going to store all the links
sunglass_opening_urls = []
url= driver.find_elements_by_xpath('//nav[@class= "yFHi8N"]//a')
for i in url[0:3]:
    sunglass_opening_urls.append(i.get_attribute('href')) #to retrieve any link we use get_attribute
```

```python
sunglass_opening_urls
```

```python
sunglass_brand=[]
for i in sunglass_opening_urls:
    driver.get(i)
    try:
        brands = driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
        for i in brands:
            sunglass_brand.append(i.text)
    except:
        pass
            
sunglass_brand
```

```python
sunglass_description=[]
for i in sunglass_opening_urls:
    driver.get(i)
    try:
        description = driver.find_elements_by_xpath("//a[@class='IRpwTa']")
        for i in description:
            sunglass_description.append(i.text)
    except:
        pass
            
sunglass_description
```

```python
sunglass_price=[]
for i in sunglass_opening_urls:
    driver.get(i)
    try:
        price = driver.find_elements_by_xpath("//div[@class='_30jeq3']")
        for i in price:
            sunglass_price.append(i.text)
    except:
        pass
            
sunglass_price
```

```python
sunglass_discount=[]
for i in sunglass_opening_urls:
    driver.get(i)
    try:
        discount = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
        for i in discount:
            sunglass_discount.append(i.text)
    except:
        pass
            
sunglass_discount
```

```python
print(len(sunglass_brand),len(sunglass_description),len(sunglass_price),len(sunglass_discount))
```

```python
data = list(zip(sunglass_description[0:100],sunglass_price[0:100],sunglass_discount[0:100]))

import pandas as pd
df = pd.DataFrame(data,columns=['sunglass_description','sunglass_price','sunglass_discount'])

df
```
